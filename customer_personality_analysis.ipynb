```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Step 1: Import the CSV file into a DataFrame
file_path = 'Test.csv'  # Replace with your actual file path
data = pd.read_csv(file_path)
```

```python
# Display the first few rows of the DataFrame
print(data.head())
```

```python
# Step 2: Summary Statistics for Numerical Columns
summary_stats = data.describe()
print(summary_stats)
```

```python
# Plot summary statistics
plt.figure(figsize=(12, 8))
sns.heatmap(summary_stats.T, annot=True, cmap="coolwarm", linewidths=0.5, fmt=".2f")
plt.title('Summary Statistics for Numerical Columns')
plt.show()
```

```python
# Step 3: Missing Values Analysis
missing_values = data.isnull().sum()
print(missing_values)
```

```python
# Plot missing values
plt.figure(figsize=(12, 4))
sns.barplot(x=missing_values.index, y=missing_values.values)
plt.title('Missing Values Analysis')
plt.xlabel('Columns')
plt.ylabel('Number of Missing Values')
plt.xticks(rotation=45)
plt.show()
```

```python
# Step 4: Sales Trends Based on Various Factors (Using Item_MRP as proxy for sales)
plt.figure(figsize=(12, 4))
sns.boxplot(data=data, x='Outlet_Size', y='Item_MRP')
plt.title('Sales Trends by Outlet Size')
plt.xlabel('Outlet Size')
plt.ylabel('Item MRP')
plt.show()



```

```python
plt.figure(figsize=(12, 4))
sns.boxplot(data=data, x='Outlet_Type', y='Item_MRP')
plt.title('Sales Trends by Outlet Type')
plt.xlabel('Outlet Type')
plt.ylabel('Item MRP')
plt.xticks(rotation=45)
plt.show()

```

```python
plt.figure(figsize=(12, 4))
sns.boxplot(data=data, x='Outlet_Location_Type', y='Item_MRP')
plt.title('Sales Trends by Outlet Location Type')
plt.xlabel('Outlet Location Type')
plt.ylabel('Item MRP')
plt.show()
```

```python
# Step 5: Customer Preferences
plt.figure(figsize=(12, 4))
sns.boxplot(data=data, x='Item_Type', y='Item_MRP')
plt.title('Customer Preferences by Item Type')
plt.xlabel('Item Type')
plt.ylabel('Item MRP')
plt.xticks(rotation=90)
plt.show()

```

```python
plt.figure(figsize=(12, 4))
sns.boxplot(data=data, x='Item_Fat_Content', y='Item_MRP')
plt.title('Customer Preferences by Item Fat Content')
plt.xlabel('Item Fat Content')
plt.ylabel('Item MRP')
plt.show()
```

```python

# Step 6: Yearly Analysis
plt.figure(figsize=(12, 4))
sns.countplot(data=data, x='Outlet_Establishment_Year', order=data['Outlet_Establishment_Year'].value_counts().index)
plt.title('Yearly Analysis of Outlet Establishment')
plt.xlabel('Outlet Establishment Year')
plt.ylabel('Count')
plt.show()
```
